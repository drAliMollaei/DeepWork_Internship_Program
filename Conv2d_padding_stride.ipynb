```python
import torch
import torch.nn as nn

# Define the input dimensions
in_channels = 3
out_channels = 16
kernel_size = 3
stride = 1
padding = 1

# Create an instance of nn.Conv2d
conv = nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding)

# Create a dummy input tensor
x = torch.randn(1, in_channels, 64, 64)
print(x.shape)
# Pass the input through the convolutional layer
output = conv(x)
print(x.shape)

print(output.shape)
```

```python
# Accessing the weights (kernels) of the convolutional layer
kernels = conv.weight.data

print(kernels.shape)  # Shape of the kernels

# Printing the first kernel
print(kernels[0])  # Accessing the first kernel

# Printing the second kernel
print(kernels[1])  # Accessing the second kernel

# Printing all the kernels
for i in range(out_channels):
    print(kernels[i])  # Accessing each kernel in the convolutional layer

# Printing the biases of the convolutional layer
biases = conv.bias.data
print(biases)
```
